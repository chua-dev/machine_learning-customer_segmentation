```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

from datetime import datetime
import datetime
```

# Reading & Preparing Data

```python
df = pd.read_excel('online_retails.xlsx')

# Drop observation if CustomerID is empty
df = df[df['CustomerID'].notna()]
```

```python
# Data describe, total 406829 observation and 8 variables
print(df.shape)
df.describe()
```

```python
# Check if there is any null value
df.isnull().sum()
```

```python
# Select random 50,000 from sample, 600k too much
df = df.sample(50000, random_state=42)
print(df.shape)
print(df.describe())
df.isnull().sum()
```

```python
# Explore head(5) of the data
df.head()
```

# Data Preprocessing
## Create RFM(Recency, Frequency, Monetary) Table

```python
print(df['InvoiceDate'].head())

# Reformat the invoice data to intended format which is date only
df['InvoiceDate'] = df['InvoiceDate'].dt.date

print(df['InvoiceDate'].head())
```

```python
# Create TotalSum by UnitPrice * Quantity
df['TotalSum'] = df['Quantity'] * df['UnitPrice']
df['TotalSum']
```

```python
# Create date variable that records recency, get the most recent date of purchase
snapshot_date = max(df.InvoiceDate) + datetime.timedelta(days=1)
print(snapshot_date)
```

```python
# Group by CustomerID to show each unique customer (their mean, min, max, count etc, for Quantity, UnitPrice, TotalSum)
df_groupby_customer = df.groupby(['CustomerID'])
df_groupby_customer.describe()
```

```python
# Create a new table out of groupby data, by indicating desired variables
customer_rfm = df_groupby_customer.agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,
    'InvoiceNo': 'count',
    'TotalSum': 'sum'
})
```

```python
# Check the newly created RFM customer DF
customer_rfm
```

```python
# Rename Columns for the customer_rfm
customer_rfm.rename(columns = {
    'InvoiceDate': 'Recency',
    'InvoiceNo': 'Frequency',
    'TotalSum': 'MoneySpent',
}, inplace=True)

customer_rfm
```

```python
import seaborn as sns

# Plot the distribution of FRM variable, check skewness
fig, ax = plt.subplots(1, 3, figsize=(15,3))
sns.distplot(customer_rfm['Recency'], ax=ax[0]) # First subplot
sns.distplot(customer_rfm['Frequency'], ax=ax[1]) # Second subplot
sns.distplot(customer_rfm['MoneySpent'], ax=ax[2]) # Third Subplot
plt.tight_layout()
plt.show()
```

```python
# Analyze the skewness of origin & 3 different transformation applied

from scipy import stats
def analyze_skewness(dataset, var):
    fig, ax = plt.subplots(2, 2, figsize=(15,7))
    sns.distplot(dataset[var], ax=ax[0,0])
    sns.distplot(np.log(dataset[var]), ax=ax[0,1])
    sns.distplot(np.sqrt(dataset[var]), ax=ax[1,0])
    sns.distplot(stats.boxcox(dataset[var])[0], ax=ax[1,1]) # Box cox power transformation
    plt.tight_layout()
    plt.show()
    
    print(dataset[var].skew().round(2))
    print(np.log(dataset[var]).skew().round(2))
    print(np.sqrt(dataset[var]).skew().round(2))
    print(pd.Series(stats.boxcox(dataset[var])[0]).skew().round(2))

# Analyzing Recency
analyze_skewness(customer_rfm, 'Recency')
```

```python
# Analyzing Frequency to choose the best transformation method to apply
analyze_skewness(customer_rfm, 'Frequency')
```

```python
# Now going for Monetary Value = Money Spent
fig, ax = plt.subplots(1, 2, figsize=(10,3))
sns.distplot(customer_rfm['MoneySpent'], ax=ax[0])
sns.distplot(np.cbrt(customer_rfm['MoneySpent']), ax=ax[1])
plt.show()

print(customer_rfm['MoneySpent'].skew().round(2))
print(np.cbrt(customer_rfm['MoneySpent']).skew().round(2))
```

```python
# Test to apply cube root in Money Spent columns, and get its values
pd.Series(np.cbrt(customer_rfm['MoneySpent'])).values
```

```python
# Applying the transformation and create new DataFrame

customer_tf = pd.DataFrame()
customer_tf['Recency'] = stats.boxcox(customer_rfm['Recency'])[0]
customer_tf['Frequency'] = stats.boxcox(customer_rfm['Frequency'])[0]
customer_tf['MonetaryValue'] = pd.Series(np.cbrt(customer_rfm['MoneySpent'])).values

# Show the newly created video
customer_tf.head()
```

# Dataset Scaling

```python
# pip install scikit-learn
from sklearn.preprocessing import StandardScaler
std_scaler = StandardScaler()

# Fit and transform the customer_tf dataset
std_scaler.fit(customer_tf)
customer_scaled = std_scaler.transform(customer_tf) # After scaled just normal Array
print(customer_scaled.mean(axis = 0).round(2))
print(customer_scaled.std(axis = 0))
```

```python
# Since it is not data frame, use panda method and display
print(customer_scaled)
pd.DataFrame(customer_scaled).head()
```

# Modelling - Chosing the optimum K-number

```python
from sklearn.cluster import KMeans

sse = {}
for n in range(1, 13): #Checking n=1 to n=12
    kmeans = KMeans(n_clusters=n, random_state=42)
    kmeans.fit(customer_scaled)
    sse[n] = kmeans.inertia_ # Inertia is the sum of squared distances of samples to their closest cluster centre.
    
plt.title('The Elbow Method')
plt.xlabel('n')
plt.ylabel('SSE')
sns.pointplot(x=list(sse.keys()), y=list(sse.values()))
plt.show()

# As we can see from number 1 and 2, the increase in class dramatically reduce the SSE
# Up to 3, more and more number of n doesnt affect much about the SSE efficiency
# Thus 3 is selected, 4 can be considered too
```

```python
# Labelling each customer into 0, 1, 2, Total 3 clusters

model = KMeans(n_clusters=3, random_state=42)
model.fit(customer_scaled)
print(model.labels_.shape)
model.labels_ 
```

# Clustering Analysis

```python
# Create one more columns in customer_rfm table, to show its class
# We still save rfm table,transformed and scaled is to fit into KMeans Model
customer_rfm['Cluster'] = model.labels_
customer_rfm
```

```python
customer_rfm.groupby('Cluster').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'MoneySpent': ['mean', 'count']
}).round(1)

# Cluster 1 spent less recent, less frequent, less amount - Inactive customer
# Cluster 0 recent average, frequent average, relatively less money - Normal Cust
# Cluster 2 spent more recent, super frequent, super high money - Loyal Cust
```
